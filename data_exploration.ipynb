### Visual data exploration with Matplotlib

```python
data1 = np.random.normal(0, 0.1, 1000)
data2 = np.random.normal(1, 0.4, 1000)
data3 = 2 + np.random.random(1000) * np.linspace(1, 5, 1000)
data4 = np.random.normal(3, 0.2, 1000)+ 0.3 * np.sin(np.linspace(0, 20, 1000))
```

```python
data = np.vstack([data1, data2, data3, data4]).transpose()
df = pd.DataFrame(data, columns = ['data1', 'data2', 'data3', 'data4'])
df.head()
```

### Line plot

```python
df.plot(title='Line Plot')
```

```python
plt.plot(df)
plt.title('Line Plot')
plt.legend(["data1", "data2", "data3", "data4"])
```

### Scatter plot

```python
df.plot(style='.')
```

```python
_ = df.plot(kind='scatter', x = 'data1', y = 'data2',
           xlim=(-1.5, 1.5), ylim=(0,3))
```

### Histogram

```python
df.plot(kind='hist',
       bins=50,
       title='Histogram',
       alpha=0.5)
```

### Cummulative plot

```python
df.plot(kind='hist',
        bins=100,
        title='Cumm distribution',
        normed=True,
        cumulative=True,
        alpha=0.2)
```

### Blox plot

```python
fig, ax = plt.subplots(2, 2, figsize=(8,8))

df.plot(ax=ax[0][0],
       title='Line plot')

df.plot(ax=ax[0][1],
       style='o',
       title = 'Scatter plot')

df.plot(ax=ax[1][0],
       kind='hist',
       bins=50,
       title='Histogram')

df.plot(ax=ax[1][1],
       kind='box',
       title='BloxPlot')

plt.tight_layout()
```

### Pie charts

```python
gt01 = df['data1'] > 0.1
piecounts = gt01.value_counts()
piecounts
```

```python
piecounts.plot(kind='pie',
              figsize=(5,5),
              explode=[0, 0.15],
              labels = ['<= 0.1', '> 0.1'],
              autopct='%1.1f%%',
              shadow=True,
              startangle=90,
              fontsize=16)
```
